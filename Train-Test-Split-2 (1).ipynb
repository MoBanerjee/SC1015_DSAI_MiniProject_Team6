```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
maindataset = pd.read_csv('Final Datasets/dataCleaned.csv')
```

```python
maindataset.head()
```

```python

#Creating a new dataset X by dropping the popularity column
X = maindataset.drop(['popularity'], axis=1)

#Creating a new dataset Y with the Popularity column
Y = maindataset['popularity']
```

```python
X
```

```python
Y
```

## Splitting the dataset into train and test datasets

```python
#Dividing the dataset into train and test dataset - 80% to Train the model and 20% to test the model
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
X_train['popularity'] = Y_train
X_test['popularity'] = Y_test
```

```python
#creating csv files for the train and test data
X_train.to_csv('Test Train Data/Train.csv')
X_test.to_csv('Test Train Data/Test.csv')
```
