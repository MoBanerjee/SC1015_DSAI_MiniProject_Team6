popularity is popularity score range which ranges from 0 to 100, with 100 being the most popular. It is calculated based on the number of streams and how recent those streams are. - categorical

duration_ms is the duration of the song given in milli seconds. - numerical

explicit indicates whether a track contains explicit content (lyrics or themes that may be unsuitable for children  or sensitive listeners). The value of this column is binary, with 1 indicating that the track contains explicit content, and 0 indicating that it does not. - categorical

energy is a measure of how powerful and energetic a song sounds.The energy value ranges from 0 to 1, with higher values indicating more energetic songs. - numerical

The key values in the Spotify dataset are assigned based on the standard pitch class notation system, which assigns a number between 0 and 11 to represent the 12 different pitch classes in western music. In the dataset, the key column contains integer values ranging from 0 to 11, where each value corresponds to a specific pitch class - numerical

loudness column represents the overall loudness of a track in decibels (dB). The loudness values are represented as floating-point numbers ranging from -60 dB to 0 dB. - numerical

mode column in the Spotify tracks dataset represents the modality of a track, i.e., whether a track is in a major or minor key. The mode is indicated by a binary value, where 1 represents a major key and 0 represents a minor key. - numerical

"speechiness" values range from 0.0 to 1.0, with higher values indicating that a track is more likely to be mostly spoken words, while lower values indicate that the track is likely to be more instrumental or music-focused.- numerical

acousticness is represented as a value between 0.0 and 1.0, where 0.0 indicates a high degree of electronic sounds and 1.0 indicates a high degree of acoustic sounds. - numerical

instrumentalness is represented as a value between 0.0 and 1.0, where a higher value indicates that the track is more instrumental and contains fewer vocals. - numerical

The "liveness" column in the Spotify dataset refers to the probability that a track was performed live. The value is represented as a float between 0.0 and 1.0, where a higher value indicates that the track is more likely to have been performed live. - numerical

valence column is a numeric indicator of the musical positiveness conveyed by a track. The values range from 0.0 to 1.0, where tracks with a higher valence value sound more positive or happy, and those with a lower valence value sound more negative or sad. - numerical

tempo values range from around 40 BPM to over 200 BPM, with an average tempo of around 120 BPM.A higher tempo generally indicates a faster-paced song, while a lower tempo suggests a slower, more relaxed song. - numerical

time signature values are represented as integers, where the first number in the time signature is represented by the integer value in the column. For example, a track with a time signature of 4/4 would have a value of 4 in the time signature column. - numerical

# Merging the Datasets

```python
merged = pd.merge(stats_data, lyrics, on='name')
```

We are merging the two cleaned datasets using the 'name' column because we want to study and compare the influence of the lyrics of the song and the other features like danceability, energy etc. on its popularity.

```python
merged.head()
```

```python
merged.info()
```

### Data Scaling

```python
#making a new dataset with only numerical data
comp_data = merged.select_dtypes(include=['float64','int64'])
comp_data.info() 

#making a new dataset with only non numeric data
nn_data = merged[[ 'release_date','explicit', 'name','Lyric', 'language']]
```

```python
# Min-Max Normalization
comp_data = (comp_data-comp_data.min())/(comp_data.max()-comp_data.min())
final_data = pd.concat((nn_data,comp_data), 1)

final_data.head(5)
```

```python
final_data = final_data.reset_index(drop=True)
```
