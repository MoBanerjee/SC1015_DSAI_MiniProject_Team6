```python
numerical_corr = {}
for i in data_numerical.columns:
    numerical_corr[i] = data_numerical[i].corr(data_numerical['popularity'])
```

### Correlation between Popularity and Numerical values

These are the **correlation values** which vary from -1 to 1.
The **correlation coefficient** is a statistical measure that **ranges between -1 and +1**, where **-1 indicates a perfect negative correlation** (i.e., when one variable increases, the other decreases), **+1 indicates a perfect positive correlation** (i.e., when one variable increases, the other also increases), and **0 indicates no correlation** (i.e., the variables are independent of each other).

```python
numerical_corr
```

```python
plt.figure(figsize=(20,20))
sb.heatmap(data_numerical.corr(), annot=True)
```

This is the **heatmap** showing the **correlation values which we calculated earlier visually**. The **lighter the colour of the box, the higher is the positive correlation** between two variables. **Darker colours** signify **stronger negative correlation**

# Natural Language Processing EDA - Lyrics

```python
import re
import nltk
from nltk.corpus import stopwords
from collections import Counter
from nltk.stem import SnowballStemmer
```

```python
# Downloading the nltk stopwords. "nltk.download()" was not working so I had to use the following code to download the stopwords.

import ssl

try:
    _create_unverified_https_context = ssl._create_unverified_context
except AttributeError:
    pass
else:
    ssl._create_default_https_context = _create_unverified_https_context

nltk.download()
```

```python
#stopwords.words('english') initializes a set of stopwords in English using the NLTK library.
stopwords = stopwords.words('english')

#initializes a stemming object that uses the Snowball algorithm for stemming
stemmer = SnowballStemmer('english')

def clean_text(text):
    #converts all the text to lowercase to ensure consistency in processing.
    text = text.lower()
    
    #removes all non-alphabetic characters and whitespace from the text string using regular expressions. The re.I|re.A flag makes the search case-insensitive and ASCII-only.
    text = re.sub(r'[^a-zA-Z\s]', '', text, re.I|re.A)
    tokens = text.split()
    
    #applies stemming to each token and removes any stopwords from the resulting list of stemmed tokens. The list comprehension here performs the stemming and filtering operations in one line.
    tokens = [stemmer.stem(token) for token in tokens if token not in stopwords]
    text = ' '.join(tokens)
    
    #returns the cleaned text string.
    return text
```

### Text- Cleaning Function Description:
Here, we are performing text cleaning function that uses the **Natural Language Toolkit (NLTK) library** in Python to preprocess a given text string.

* **Stopwords** are common words that are usually removed from text data during preprocessing because they do not carry much meaning.

* **Stemming** is the process of reducing a word to its base form (or stem) by removing affixes (prefixes, suffixes, etc.). 

* This function also involves **tokenization** that splits the text string into individual words or tokens.

* Overall, this function is a simple yet effective way to preprocess text data by removing non-alphabetic characters, stopwords, and reducing words to their base forms. This can help improve the quality of the text data for further analysis.

```python
dataset['Lyric Altered'] = dataset['Lyric'].apply(lambda x: clean_text(x))
```

```python
dataset['lenth'] = dataset['Lyric Altered'].apply(lambda x: len(x))
sb.boxplot(x=dataset['lenth'])
```

Here, we are creating a **new column** in the dataset called **lenth** with the **length of each element x of the 'Lyric Altered' column**. After creating the column, we have performed EDA by creating a **box plot**. 

### Unique Songs

```python
#Unique songs

unique_songs = len(dataset['Lyric Altered'].unique())
print("Number of unique songs: ", unique_songs)

#Distribution of popularity index
sb.set(style="whitegrid")
plt.figure(figsize=(20,10))
sb.histplot(dataset['popularity'], kde=True)
plt.title('Distribution of song popularity')
plt.xlabel('Popularity Index')
plt.ylabel('Frequency')
```

Here we have plotted the **histogram plot** along with the **kernel density estimate** of **popularity index**.

```python
#counting the frequency of elements in the lyrics

word_counter = Counter()

for lyrics in dataset['Lyric Altered']:
    word_counter.update(lyrics.split())
```

```python
#finding the top 20 common words

top_20_words = word_counter.most_common(20)
top_words, top_counts = zip(*top_20_words)
```

```python
plt.figure(figsize=(20,10))
plt.bar(top_words, top_counts)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 20 words in the lyrics')
plt.xticks(rotation=45)

plt.show()
```

This is a **bar plot** of the **top 20 common words in the lyrics**
